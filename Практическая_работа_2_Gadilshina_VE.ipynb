**Задание 2.5**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2,	17.6,	8.8,	8,	7.7,	6.7]
ax.bar(languages, popularity)

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth= '0.5', color='red')

plt.grid(which='minor', linestyle=':', linewidth= '0.5', color='black')

for language, pop in zip(languages, popularity):
    ax.text(language, pop + 0.3, f'{pop:.1f}', ha='center', va='bottom')

plt.legend()
plt.title("Popularity of Programming Language Worldwide, Oct 2017 compared to a year ago")
plt.xlabel("Languages")
plt.ylabel("Popularity")

plt.show()
```

**Часть 2. Провести анализ датасетов с помощью библиотеки Matplotlib, Pyplot:**

Этот набор данных [Lifestyle_and_Wellbeing_Data](https://www.kaggle.com/datasets/ydalat/lifestyle-and-wellbeing-data/data) содержит 15 972 ответов на опросы с 24 характеристиками, описывающими, как мы живем.

1. **Timestamp**: Дата прхождения опроса
2. **FRUITS_VEGGIES**: HOW MANY FRUITS OR VEGETABLES DO YOU EAT EVERYDAY? От 0 до 5. Select 5 if more than 5.
3. **DAILY_STRESS**: HOW MUCH STRESS DO YOU TYPICALLY EXPERIENCE EVERYDAY? От 0 до 5.
4. **PLACES_VISITED:** HOW MANY NEW PLACES DO YOU VISIT? От 0 до 10, за 12 месяцев.
5. **CORE_CIRCLE:** HOW MANY PEOPLE ARE VERY CLOSE TO YOU? От 0 до 10.
6. **SUPPORTING_OTHERS:** HOW MANY PEOPLE DO YOU HELP ACHIEVE A BETTER LIFE? От 0 до 10 за последние 12 месяцев.
7. **SOCIAL_NETWORK:** WITH HOW MANY PEOPLE DO YOU INTERACT WITH DURING A TYPICAL DAY? От 0 до 10.
8. **ACHIEVEMENT:** Over the last 12 months, personal achievements known to your family, close friends or co-workers. От 0 до 10.
9. **DONATION:** Over a period of 12 months. От 0 до 5.
10. **BMI_RANGE:** Your body mass in kg divided by the square of your height in meters. Below 25 - 1 и Above 25 - 2
11. **TODO_COMPLETED:** HOW WELL DO YOU COMPLETE YOUR WEEKLY TO-DO LISTS?  On a scale of 0 = not at all to 10 = very well.
12. **FLOW:** IN A TYPICAL DAY, HOW MANY HOURS DO YOU EXPERIENCE "FLOW"? «Поток» определяется как психическое состояние, в котором вы полностью погружены в выполнение какой-либо деятельности. Затем вы испытываете чувство энергичной сосредоточенности, полной вовлеченности и удовольствия в процессе этой деятельности. От 0 до 10.
13. **DAILY_STEPS:** HOW MANY STEPS (IN THOUSANDS) DO YOU TYPICALLY WALK EVERYDAY? От 0 до 10.
14. **LIVE_VISION:** FOR HOW MANY YEARS AHEAD IS YOUR LIFE VISION VERY CLEAR FOR? От 0 до 10 в годах.
15. **SLEEP_HOURS:** ABOUT HOW LONG DO YOU TYPICALLY SLEEP? От 0 до 10.
16. **LOST_VACATION:** Unused vacation days, lost or carried forward into the following year. От 0 до 10.
17. **DAILY_SHOUTING:** HOW OFTEN DO YOU SHOUT OR SULK AT SOMEBODY? От 0 до 10.
18. **SUFFICIENT_INCOME:** HOW SUFFICIENT IS YOUR INCOME TO COVER BASIC LIFE EXPENSES? (Доход для жизни). Not or hardly sufficient - 1, Sufficient - 2.
19. **PERSONAL_AWARDS:** HOW MANY RECOGNITIONS HAVE YOU RECEIVED IN YOUR LIFE? От 0 до 10.
20. **TIME_FOR_PASSION:** HOW MANY HOURS DO YOU SPEND EVERYDAY DOING WHAT YOU ARE PASSIONATE ABOUT? От 0 до 10.
21. **WEEKLY_MEDITATION**: IN A TYPICAL WEEK, HOW MANY TIMES DO YOU HAVE THE OPPORTUNITY TO THINK ABOUT YOURSELF? От 0 до 10.
22. **AGE**: Less than 20, 21 to 35, 36 to 50, 51 or more.
23. **GENDER**: Female, Male.
24. **WORK_LIFE_BALANCE_SCORE:** Баланс баланса между работой и личной жизнью представляет собой сумму всех пяти категорий: плохой балл — ниже 550, хороший — выше 680, отличный — выше 700.

**DAILY_STRESS** - object, следовательно нужна обработка данных (удаление, очистка, преобразование в int).

```python
df_3 = df_1[df_1['DAILY_STRESS'].str.isnumeric()] #удаление строки с некорректным значением

df_3['DAILY_STRESS'] = df_3['DAILY_STRESS'].astype(int) #преобразование в тип int

df_3.info()
```

```python
df_3.describe()
```

**1)	построить	несколько	графиков	на	одном	поле**

```python
figure, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].hist(df_3['FRUITS_VEGGIES'], bins=5, edgecolor='black')
axs[0].set_xlabel('FRUITS_VEGGIES')
axs[0].set_ylabel('Количество респондентов')
axs[0].set_title('Распределение съедаемого количества овощей и фруктов в день')

axs[1].hist(df_3['DAILY_STRESS'], bins=5, edgecolor='black')
axs[1].set_xlabel('DAILY_STRESS')
axs[1].set_ylabel('Количество респондентов')
axs[1].set_title('Распределение степени ежедневного стресса')

plt.tight_layout()
plt.show()
```

**2)	построить	диаграммы	для	категориальных	данных**

Категориальные данные:
1. **SUFFICIENT_INCOME:** HOW SUFFICIENT IS YOUR INCOME TO COVER BASIC LIFE EXPENSES? (Доход для жизни). Not or hardly sufficient - 1, Sufficient - 2.
2. **BMI_RANGE:** Your body mass in kg divided by the square of your height in meters. Below 25 - 1 и Above 25 - 2.
3. **AGE:** Less than 20, 21 to 35, 36 to 50, 51 or more.
4. **GENDER:** Female, Male.                     

```python
df_3['SUFFICIENT_INCOME'] = df_3['SUFFICIENT_INCOME'].replace({1: 'Not or hardly sufficient', 2: 'Sufficient'})
plt.figure(figsize=(6, 3))
sns.countplot(data=df_3, x="SUFFICIENT_INCOME")
plt.xlabel("Достаточность дохода")
plt.ylabel("Количество респондентов")
plt.title("Распределение по достаточности дохода на повседневную жизнь")
plt.show()
```

```python
df_3['BMI_RANGE'] = df_3['BMI_RANGE'].replace({1: 'Below 25', 2: 'Above 25'})
plt.figure(figsize=(6, 3))
sns.countplot(data=df_3, x="BMI_RANGE")
plt.xlabel("Индекс массы тела")
plt.ylabel("Количество респондентов")
plt.title("Распределение по Индексу массы тела")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df_3, x="AGE")
plt.xlabel("Возрастная категория")
plt.ylabel("Количество респондентов")
plt.title("Распределение по возрастным категориям")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df_3, x="GENDER")
plt.xlabel("Пол")
plt.ylabel("Количество респондентов")
plt.title("Соотношение между мужчинами и женщинами")
plt.show()
```

**3)	построить	столбчатые	и	круговые	диаграммы**

```python
gender_counts = df_3['GENDER'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.2f%%', startangle=100)
plt.title("Соотношение между мужчинами и женщинами")
plt.axis('equal')
plt.show()
```

```python
bins = [0, 549, 679, 699, float('inf')]
labels = ['Плохой', 'Средний', 'Хороший', 'Отличный']
df_3['SCORE_CATEGORY'] = pd.cut(df_3['WORK_LIFE_BALANCE_SCORE'], bins=bins, labels=labels)

plt.figure(figsize=(8, 6))
df_3['SCORE_CATEGORY'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Баланс между работой и личной жизнью')
plt.xlabel('Уровень баланса между работой и личной жизнью ')
plt.xticks(rotation=45)
plt.ylabel('Количество респондентов')

plt.show()
```

**5)	построить	3D-график**

```python
from mpl_toolkits.mplot3d import Axes3D

x = df_3['PLACES_VISITED']
y = df_3['SUPPORTING_OTHERS']
z = df_3['WORK_LIFE_BALANCE_SCORE']

fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='red', marker='o')

ax.set_xlabel('PLACES_VISITED')
ax.set_ylabel('SUPPORTING_OTHERS')
ax.set_zlabel('WORK_LIFE_BALANCE_SCORE')

plt.title('3D-график')

plt.show()
```

```python
x = df_3['PLACES_VISITED']
y = df_3['SUPPORTING_OTHERS']
z = df_3['WORK_LIFE_BALANCE_SCORE']

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_trisurf(x, y, z, cmap='viridis')

ax.set_xlabel('PLACES_VISITED')
ax.set_ylabel('SUPPORTING_OTHERS')
ax.set_zlabel('WORK_LIFE_BALANCE_SCORE')

plt.title('Поверхность WORK_LIFE_BALANCE_SCORE от PLACES_VISITED и SUPPORTING_OTHERS')

plt.show()
```
