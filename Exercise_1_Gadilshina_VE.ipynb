### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
```

### Step 4. Which continent drinks more beer on average?

```python
#Group
continent_and_beer = drinks.groupby('continent')['beer_servings'].mean()
```

```python
average_beer_drinks = continent_and_beer.max()
continent_and_beer
```

### Step 5. For each continent print the statistics for wine consumption.

```python
wine_by_continent = drinks.groupby('continent')['wine_servings'].describe()
wine_by_continent
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
mean_alcohol_consumption_per_continent = drinks.groupby('continent').mean()
mean_alcohol_consumption_per_continent
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
median_alcohol_consumption_per_continent = drinks.groupby('continent').median()
median_alcohol_consumption_per_continent
```
